```python
from ultralytics import YOLO

# Load the trained YOLO model
model = YOLO("runs/detect/train8/weights/best.pt")  # Update the path to your model weights
```

```python
from PIL import Image
from IPython.display import Image as IPImage

# Path to the WEBP file
webp_path = "/content/OIP (1).jpg"

# Convert the WEBP image to PNG
converted_path = "converted_image.png"
image = Image.open(webp_path)
image.save(converted_path, format="PNG")

# Display the PNG image
IPImage(filename=converted_path, width=600)
```

```python
# Make predictions on the uploaded image
results = model.predict(source=uploaded_file, save=True)

# Display the first result (as a list of results is returned)
results[0].show()  # This accesses the first result and shows the image with labels

# Optionally, save the results to a file and provide the download link
print(f"Predictions saved in: {results[0].save_dir}")
```
